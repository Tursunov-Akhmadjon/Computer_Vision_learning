```python
import cv2
from PIL  import Image,ImageEnhance,ImageFilter,ImageDraw,ImageDraw2,ImageOps
import matplotlib.pyplot as plt


img_path='/content/drive/MyDrive/Colab Notebooks/Computer Vision/Face_anonymization_/person.jpg'

img=Image.open(img_path)


# Enhance
enhancer=ImageEnhance.Brightness(img)
img_bright=enhancer.enhance(1.5)

# # Blur
img_blur=img.filter(ImageFilter.GaussianBlur(10))

# # Draw
img_rect = img.copy()
x=ImageDraw.Draw(img_rect)
x.rectangle([70,30,120,100],outline='blue',width=2)

# #Flip
img_flip=ImageOps.flip(img)

# #Resize
img_resized=img.resize((300,400))

images=[img, img_bright, img_blur, img_rect, img_flip, img_resized]

# Display the image using matplotlib.pyplot
# Create a figure with 3 subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 5))

# Display each image in a subplot
for i, ax in enumerate(axes.flat):  # Iterate through subplots
    ax.imshow(images[i])
    ax.axis('off')  # Hide axes
    ax.set_title(f'Image {i+1}')  # Set title

# Show the plot
plt.show()
```
