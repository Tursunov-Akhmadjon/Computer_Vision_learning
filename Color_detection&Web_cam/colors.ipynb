```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "photo.jpg"  #image path
image = cv2.imread(image_path) # variable for image

if image is None: # Get error if image is not found
    print("Error: Image not found or cannot be loaded. Check the file path.")

else:
    
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert OpenCV BGR format to RGB
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)  # Convert OpenCV BGR format to HSV
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)  # Convert OpenCV BGR format to LAB
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert OpenCV BGR format to GRAY

    # Create subplots for displaying images
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))

    # Display RGB Image
    axes[0].imshow(rgb_image)
    axes[0].set_title("RGB Image")
    axes[0].axis("off")

    # Display HSV Image
    axes[1].imshow(hsv_image)
    axes[1].set_title("HSV Image")
    axes[1].axis("off")

    # Display LAB Image
    axes[2].imshow(lab_image)
    axes[2].set_title("LAB Image")
    axes[2].axis("off")

    # Display Grayscale Image
    axes[3].imshow(grayscale_image, cmap="gray")
    axes[3].set_title("Grayscale Image")
    axes[3].axis("off")

    plt.show()
```
