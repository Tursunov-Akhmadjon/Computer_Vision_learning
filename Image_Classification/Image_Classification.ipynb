```python
#Libraries

import os
import cv2
import pickle
import numpy as np

from skimage import io
from skimage.transform import resize
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dir='/content/drive/MyDrive/Colab Notebooks/Computer Vision/Image_Classification/Animals' #Dataset Folder path
classes=['cats','dogs'] #Class names
```

```python
#Creating 2 list for data and labels
data=[]
labels=[]
```

```python
#Dataset preprocessing
for classes_idx, clas in enumerate(classes):
  for file in os.listdir(os.path.join(dir,clas)):
    img_path=os.path.join(dir,clas,file)
    img=cv2.imread(img_path)
    img=resize(img,(150,150))
    data.append(img.flatten())
    labels.append(classes_idx)

data=np.asarray(data)  # x
labels=np.asarray(labels)  # y
```

```python
#Spliting dataset to 2 datasets (for train, for test)
x_train,x_test,y_train,y_test=train_test_split(data,labels,test_size=0.3,shuffle=True,stratify=labels)
```

```python
classifier=SVC()

classifier
```

```python
parametrs=[{'gamma':[0.01, 0.001, 0.0001],'C':[1, 10, 100]}]

parametrs
```

```python
grid_search=GridSearchCV(classifier,parametrs)

grid_search
```

```python
#Training
grid_search.fit(x_train,y_train)
```

```python
#Best Estimator
best_estimator=grid_search.best_estimator_

best_estimator
```

```python
#Prediction
y_pred=best_estimator.predict(x_test)

y_pred
```

```python
#Accuracy
score=accuracy_score(y_pred,y_test)

score
```
